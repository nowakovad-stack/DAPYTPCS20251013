## Příprava prostředí

```python
import os
import pandas

__file__ = __vsc_ipynb_file__  
SCRIPT_FOLDER = os.path.dirname(__file__)
PROJECT_FOLDER = os.path.dirname(SCRIPT_FOLDER)
DATA_FOLDER = os.path.join(PROJECT_FOLDER,'data')
```

```python
# nazvy sloupcu v DataFrames
COL_ID = 'pid'
COL_MODEL = 'model'
COL_BEZDPH = 'bez_dph'
COL_SDPH = 's_dph'
COL_PRODUKT = 'Produkt'
COL_NAKUP = 'cena_nakup'
COL_SKLADEM = 'skladem'
COL_PRODANO = 'prodano'
COL_SKUPINA = 'skupina'
COL_HODNOTA = 'hodnota_na_sklade'
COL_VYNOS = 'vynos'
COL_ZISK = 'zisk'
COL_MARZE = 'marze'
```

## Čtení kosmetiky ze souboru

```python
file_kosm = r"C:\Users\Dorota\Desktop\datová analytika\working\data\source\sklad\cenik_kosmetiky.csv"

kosm_cfg = dict (
    #nrows=10,
    thousands=' ',
    usecols=[0,1,2,3],
    skiprows= 1,
    header= None,
    names=[COL_ID,COL_MODEL,COL_BEZDPH,COL_SDPH],
    index_col=[0]
    )

kosmetika = pandas.read_csv(file_kosm,**kosm_cfg)
kosmetika.info()
kosmetika
```

## Výběr sloupců

### Výběr jménem

```python
kosmetika[COL_MODEL]
kosmetika[[COL_MODEL,COL_SDPH]]#.info()
```

### Výběr pozicí

```python
kosmetika.iloc[3:7]
kosmetika.iloc[3:7,1:3] #omezí rádky a následně sloupce
```

```python
sdph = kosmetika.iloc[3:20][[COL_MODEL,COL_SDPH]]
sdph.info()
sdph
```

### Operace se sloupci

```python
kosmetika[COL_SKLADEM]= 10 #konstanta
kosmetika["dph"]= kosmetika[COL_SDPH] - kosmetika[COL_BEZDPH] #vazba mezi sloupce
kosmetika[COL_PRODANO] = [ii * 2 for ii in range(0,len(kosmetika))] #list nových hodnot

kosmetika
```

## Podmínečný výběr řádků
`data[<podminka>]` nebo `data[<podminka>][<seznam_sloupců>]`

### Podmínka numericky

```python
kosmetika[kosmetika[COL_PRODANO] < 20] #numerická podmínka
kosmetika[(kosmetika[COL_PRODANO] < 20)&(kosmetika[COL_BEZDPH] < 150)]#[[COL_MODEL,COL_SDPH]] #kombinace dvou podmínek s "&"
```

### Textové podmínky

```python
kosmetika[kosmetika[COL_MODEL].str.contains("PET")] #textová podmínka na část řetězce
kosmetika[kosmetika[COL_MODEL].str.contains("pet")] #textová podmínka na část řetězce "malým"

kosmetika[kosmetika[COL_MODEL].str.contains("PET",case=False)] #ignoruje velikost písmen
```

```python
kosmetika[kosmetika[COL_MODEL].str.contains(r"\bPET\b",case=False)] #regulární výrazy, textová filtrace
```

```python
kosmetika[kosmetika[COL_MODEL].str.contains(r"\bPET\b|\bsprej\b",case=False)] #regulární výrazy - více podmínek spojujeme uprostřed uvozovek "|"
```

## Řazení podle sloupce

```python
kosmetika.sort_values(COL_BEZDPH)
kosmetika.sort_values(COL_BEZDPH,ascending=False)
kosmetika.sort_values([COL_BEZDPH,COL_PRODANO],ascending=[False,True])#dle více sloupců, musí bžt v listu a řazení lze taky "vrstvit" jako list
```

```python
kosmetika.sort_index()
```

```python
kosort = kosmetika.sort_values([COL_BEZDPH,COL_PRODANO],ascending=[False,True])

kosort.head(10) #top 10 sle seřazení
kosort.tail(10) #vybere spodních 10 dle seřazení

pandas.concat([kosort.head(10),kosort.tail(10)]) #spojení 2 dataframů do jedné tabulky
```
