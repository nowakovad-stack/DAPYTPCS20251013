## Čtení více `DataFrame` do `dict`

```python
MENY = ('EUR','USD','GBP')
KURZY_URL = "https://www.cnb.cz/cs/financni-trhy/devizovy-trh/kurzy-devizoveho-trhu/kurzy-devizoveho-trhu/vybrane.txt?od=01.10.2025&do=31.10.2025&mena={mena}&format=txt"

kurzy = {

}

for mena in MENY:
    url = KURZY_URL.format(mena=mena)
    print(f"cteme kurzy '[mena]' z:", url)

    df = pandas.read_table(url,**cnb_cfg)
    kurzy[mena] = df

print("stahli jsme si kurzy:",kurzy.keys())
```

```python
kurzy.get('GBP')
```

## Uložení `DataFrame` do souboru

### CSV

```python
CSV = os.path.join(DATA_FOLDER, 'working', "kurzy_{mena}.csv")

for mena, df in kurzy.items():
    df.to_csv(CSV.format(mena=mena, index=False))
```

```python
XLS = os.path.join(DATA_FOLDER, 'working', "kurzy.xlsx")
with pandas.ExcelWriter(XLS) as xlswr:
    for mena, df in kurzy.items():
        df.to_excel(xlswr, index=False,sheet_name=mena)
```
