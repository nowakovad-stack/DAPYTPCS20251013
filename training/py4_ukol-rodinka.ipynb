# Úkol: Rodinka

Navrhněte datovou strukturu, ve které popíšete rodinné vztahy:
- každá osoba bude v proměnné typu `dict` a musí mít jméno `jmeno`, další informace jsou volitelné
- minimálně tři generace
- v každé generaci minimálně dvě osoby
- rodiče jsou vzájemně propojeni
- rodiče mají vazbu na své potomky

## Import

```python
import yaml
```

## Definice konstant


```python
KEY_NAME ='jmeno'
KEY_PARTNER = 'partner'
KEY_CHILDREN = 'deti'
KEY_EXTRAS = 'zajimavost'
```

## Datová struktura

osoby /uložené v proměnných/

```python
anna = {KEY_NAME: 'Anna Nova'}
boris = {KEY_NAME: 'Boris Novy'}
cecilie = {KEY_NAME: 'Cecilie Nova'}
david = {KEY_NAME: 'David Stary'}
filip = {KEY_NAME: 'Filip stary'}
gustav = {KEY_NAME: 'Gustav Stary'}
```

`rodina =` seznam všech osob

```python
rodina = [anna, boris, cecilie, david, filip, gustav]
print(rodina)
```

vazby ... definice vazeb mezi členy rodiny

```python
anna[KEY_PARTNER] = boris
anna[KEY_CHILDREN] = [cecilie]
boris [KEY_PARTNER] = anna
boris[KEY_CHILDREN] = [cecilie]
cecilie [KEY_PARTNER]= david
cecilie [KEY_CHILDREN] = [filip, gustav]

david [KEY_PARTNER] = cecilie
david [KEY_CHILDREN] = [filip, gustav]

filip [KEY_EXTRAS] = "dvojcata"
gustav [KEY_EXTRAS] = "dvojcata"
```

## Práce s rodinkou
cyklus `for`  
Podminka `if`

```python
for osoba in rodina:
    print(osoba.get(KEY_NAME),end=' ')
    print(f"ma partnera: {osoba.get(KEY_PARTNER,{}).get(KEY_NAME)}" if osoba.get(KEY_PARTNER) else "partnera nema") 
    if osoba.get(KEY_CHILDREN):
        print("\tma deti:", [child.get(KEY_NAME) for child in osoba.get(KEY_CHILDREN)]) 
    else:
        print("\tdeti nema")
```
