## Práce s textem "řetězce"

Pro zápis textu musíme vždy použít uvozovky **(`"`)** neho apostrof **(`'`)**

### Funkce `print()`
"Vypíše vše, co jí řekneme"

```python
print("Hello")
print('Ahoj')
print("I don't need it")
```

### Spojování textu

```python
print("Praha" + "Coding" + "School")
print("Praha" + " Coding" + " School")
print("Praha" + " " + "Coding" + " " + "School")
print("Dorota" *10)
```

### Manipulace s textem

```python
print("Praha Coding School".lower())
print("Praha Coding School".upper())
print("Praha Coding School".replace('o','A'))
print("Praha Coding School".replace(' ','').lower())
```

### Podřetězce

```python
print(len("Praha Coding School"))
print("prvni znak"[0])
print("prvni znak"[6])

print("prvních 5 znaků"[0:5])
print("posledních 5 znaků"[-5:])
print("od 5 do konce"[4:])
```

## Proměnné
**"Proměnná"** je jméno, za kterým se skrývá nějaká hodnota   
Hotnotu získává přiřazovacím příkazem `jmeno_promenne = vyraz`   
Její hodnotu získáme uvedením jejího jména ve výrazu nebo na jiném vhodném místě   
Jméno proměnné může obsahovat alfanumerické znaky a podtržítko (`_`), nesmí začínat číslem   
Doporučuje se používat vhodně popisná jména; ideálně vycházející z angličtiny

```python
a = 4
print(a)

b = 45
print(b)

print("a + b =", a+b)
print(f"{a:.2f} + {b} = {a + b}")

ROVNICE = "{} + {} - {}"
print(ROVNICE.format(a,b,a+b))
```
