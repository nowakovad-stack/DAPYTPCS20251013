## Zápis `dict` na disk
Zkopíruj si rodinku z [`py4_ukol-rodinka.ipynb`](py4_ukol-rodinka.ipynb)

```python
KEY_NAME ='jmeno'
KEY_PARTNER = 'partner'
KEY_CHILDREN = 'deti'
KEY_EXTRAS = 'zajimavost'

anna = {KEY_NAME: 'Anna Nova'}
boris = {KEY_NAME: 'Boris Novy'}
cecilie = {KEY_NAME: 'Cecilie Nova'}
david = {KEY_NAME: 'David Stary'}
filip = {KEY_NAME: 'Filip stary'}
gustav = {KEY_NAME: 'Gustav Stary'}

rodina = [anna, boris, cecilie, david, filip, gustav]

anna[KEY_PARTNER] = boris
anna[KEY_CHILDREN] = [cecilie]
boris [KEY_PARTNER] = anna
boris[KEY_CHILDREN] = [cecilie]
cecilie [KEY_PARTNER]= david
cecilie [KEY_CHILDREN] = [filip, gustav]

david [KEY_PARTNER] = cecilie
david [KEY_CHILDREN] = [filip, gustav]

filip [KEY_EXTRAS] = "dvojcata"
gustav [KEY_EXTRAS] = "dvojcata"

soubor = os.path.join(WORK_FOLDER,'rodinka.yml')
with open(soubor, 'w', encoding='UTF-8') as ff:
    yaml.dump(rodina,ff)
```

```python
rodina2= None
with open(soubor, 'r', encoding='UTF-8') as ff:
    rodina2 = yaml.safe_load(ff)
print(type(rodina2),len(rodina2))
print(rodina2[4])
```
